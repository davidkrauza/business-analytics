### 1. Generate descriptive statistics using Python

As with Lab 02, we will start off by importing the pandas library, but we are also going to use the iris dataset that comes with the Python. The data set consists of 50 samples from each of three species of Iris (Iris setosa, Iris virginica and Iris versicolor). Four features are measured from each sample: the length and the width of the sepals and petals, in centimetres

```python
import pandas as pd
import numpy as np
from sklearn import datasets
iris = datasets.load_iris()
df = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                       columns= iris['feature_names'] + ['target'])
df
```

The iris dataset is not stored in a pandas dataframe, so we had to create a panads dataframe with it.

```python
df['sepal length (cm)'].describe()
```

The process of displaying descriptive statistics in Python is super easy. The pandas library give us all the functions we need to generate them. As you can see from the previous code block we picked one column in the dataframe and called the describe() and Python printed the count of items, measures of central tendency, and interquartile ranges.

```python
df.groupby(by='target').describe()
```

Getting the descriptive statistics for one column is not partiuarly useful. You'll want to see them for each type of Iris (enumerated as 0 to 2 in the target column in this dataset). To look at the descriptive statistics for each type, you'll first groupby the target then call the describe() function.

```python
df['petal length (cm)'].median()
```

```python
df.groupby('target').median()
```

```python
df['petal length (cm)'].var()
```

```python
df.groupby(by='target').var()
```

```python
df.groupby(by='target').corr()
```

The last five code blocks did essentially the same thing we did with the describe function but with median and variance.

### 2. Generate data visualizations using Python

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df.groupby(by='target')['sepal length (cm)'].mean().plot(kind='bar')
```

```python
df.plot(kind='scatter', x='sepal length (cm)', y='sepal width (cm)')
```

```python
#create some data
x_series = [0,1,2,3,4,5]
y_series_1 = [x**2 for x in x_series]
y_series_2 = [x**3 for x in x_series]

#plot the two lines
plt.plot(x_series, y_series_1)
plt.plot(x_series, y_series_2)
```

```python
fig = sns.FacetGrid(df, hue='target', size=5).map(plt.scatter, 'sepal length (cm)', 'sepal width (cm)').add_legend()
```
