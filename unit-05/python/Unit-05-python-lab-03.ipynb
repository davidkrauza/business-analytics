As with Lab 02, we will start off by importing the pandas library, but we are also going to use the iris dataset that comes with the Python. The data set consists of 50 samples from each of three species of Iris (Iris setosa, Iris virginica and Iris versicolor). Four features are measured from each sample: the length and the width of the sepals and petals, in centimetres

```python
import pandas as pd
import numpy as np
from sklearn import datasets
iris = datasets.load_iris()
df = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                       columns= iris['feature_names'] + ['target'])
df
```

```python
df['sepal length (cm)'].describe()
```

```python
df.groupby(by='target').describe()
```

```python
df['petal length (cm)'].median()
```

```python
df.groupby('target').median()
```

```python
df['petal length (cm)'].var()
```

```python
df.groupby(by='target').var()
```

```python
df.groupby(by='target').corr()
```
